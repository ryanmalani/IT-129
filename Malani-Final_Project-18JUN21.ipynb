```python
#data vizualization (chart)
import plotly.graph_objs as go
import plotly.express as px
```

Creating Volume Profile Indicator:

```python
def volume_profile(df): #takes a dataframe as a parameter  
    sorted_df = df.sort_values('Volume', ascending=True)
    poc_vol = sorted_df['Volume'].idxmax() #finding the index to the largest volume instance in the dataframe to be the point of control
    poc_df = sorted_df.loc[poc_vol] #returns the row of the point of control
    poc_price = poc_df.iloc[1] #returns the adjusted close price at the poc
    #print(poc_price)
    
    #close_price = df['Close'].to_numpy()
    #volume = df['Volume'].to_numpy()
    
    #creating the candlestick chart figure
    chart = go.Figure()
    
    #adding the candlesticks to the graph
    chart.add_trace(
        go.Candlestick(
            x=df.index,
            open=df.Open,
            high=df.High,
            low=df.Low,
            close=df.Close))
    
    #adding the point of control as a line to the chart
    chart.add_hline(y=poc_price)
    
    #adding titles to the chart
    chart.update_layout(
        title_text='Candlestick Chart',
        yaxis_title='Price (USD/Share)',
        xaxis_title='Date/Time')
    
    #showing the candlestick chart
    chart.show()
    
    #creating a histogram for volume profile
    vol_pro = px.histogram(df, x='Volume', y='Close', nbins=150, orientation='h')
    vol_pro.show()
```

Acquiring the data using input and the Yahoo! Finance API

```python
def get_data():
    #getting user input for API arguments
    tick = input('Enter Ticker(s) (Ex: AAPL): ')
    per = input('Enter Time Period (Ex: 5d): ')
    inv = input('Enter Time Interval (Ex: 15m): ')

    #using user inputs as API arguments
    df = yf.download(tickers=tick, period=per, interval=inv) 
    df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
    return df
```

```python
data = get_data()
volume_profile(data)
```
